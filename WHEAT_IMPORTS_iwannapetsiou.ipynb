```python
df1 = df1.rename(columns={df1.columns[0]:'country_name',
                         df1.columns[1]:'country_code',
                         df1.columns[3]:'pop'})

df1
```

```python
df2 = df2.rename(columns={df2.columns[1]:'country_code',
                          df2.columns[2]:'importer',
                         df2.columns[3]:'importer_region',
                         df2.columns[6]:'value_1000usd_for_russia',
                         df2.columns[7]:'weight_1000kg_for_russia'})

df2
```

```python
df3 = df3.rename(columns={df3.columns[1]:'country_code',
                          df3.columns[2]:'importer',
                         df3.columns[3]:'importer_region',
                         df3.columns[6]:'value_1000usd_for_ukraine',
                         df3.columns[7]:'weight_1000kg_for_ukraine'})

df3
```

```python
df4 = df1[['country_code','GDP','pop']]
df4
```

```python
df5=df2[['country_code','importer','importer_region','value_1000usd_for_russia','weight_1000kg_for_russia']]
df5
```

```python
df6=df3[['country_code','importer','importer_region','value_1000usd_for_ukraine','weight_1000kg_for_ukraine']]
df6
```

```python
df7 = pd.merge(df4, df5, on='country_code', how='left')
df7
```

```python
df8 = pd.merge(df7, df6, on=['country_code','importer','importer_region'], how='left')
desired_order = ['country_code', 'importer', 'importer_region', 'value_1000usd_for_russia', 'value_1000usd_for_ukraine','weight_1000kg_for_russia','weight_1000kg_for_ukraine','GDP','pop']
df9= df8[desired_order]
df9
```

```python
df9[df9.duplicated(keep=False)]
```

```python
df9 =df9.dropna()
df9
```

```python
df9 = df9.reset_index().drop(['index'],axis=1)

df9
```

# 4. How many total tones of wheat have been imported in Europe from both Ukraine and Russia?

```python
europe_df = df9.loc[df9['importer_region'] == 'Europe']
europe_df 
europe_df['weight_1000kg_for_russia'].sum() + europe_df['weight_1000kg_for_ukraine'].sum()
```

# 5. Based on weight of total imports per capita, what are the top-10 wheat importers from both Ukraine and Russia? And what is the imported weight per capita in each top importer?

<b> <i> Present your analysis in a tiny dataframe of two columns (importer & weight_per_capita) </i></b>!

```python

df9['weight_per_capita'] = (df9['weight_1000kg_for_russia'] + df9['weight_1000kg_for_ukraine']) / df9['pop']
top_10 = df9.sort_values(by='weight_per_capita', ascending=False).head(10)

result_df = top_10[['importer', 'weight_per_capita']]
result_df.reset_index().drop(['index'],axis=1)
```

# 6. What is the total amount of money spent on wheat imports by region?

```python
df9['total_wheat_imports'] = df9['value_1000usd_for_russia'] + df9['value_1000usd_for_ukraine']

total_imports_by_region = df9.groupby('importer_region')['total_wheat_imports'].sum()
total_imports_df = total_imports_by_region.reset_index(name='Total Wheat Imports')
total_imports_df
```
